```python
import seaborn as sns
%matplotlib inline
```

```python
#visualization of the different charges in the Bronx
g=sns.displot(BX['OFNS_DESC'])
g.set_xticklabels(rotation=45)
```

In this project I compared arrests in the Bronx and Manhattan.I found that the main chargee at arrest in the Bronx and Manhattan were Assault 3 and Petit Larceny respectively. Furthermore, the age group most reperesented in arrests in both Manhattan and the Bronx is 25-44. Lastly, Black people are overrepresented in the arrests in both boroughs. Arrests do not necesarily mean convictions, so we will not know from this data just how many people were in fact guilty of the crime they were appreheneded for. This dataset may paint a picture of the NYPD being heavily preoccupied with crime in our streets, but it may well also be proof of the hypersurveillance state we live under; one that does not neessarily prioritize justice but rather punishment.
